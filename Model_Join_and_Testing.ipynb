```python
import cv2
import numpy as np
import os
from tensorflow.keras.models import load_model


print("Loading models...")

# Flood model
flood_model = load_model(
    "/content/drive/MyDrive/major/Flood_model.h5"
)

# Disaster model
disaster_model = load_model(
    "/content/drive/MyDrive/major/disaster_cnn_model.h5"
)

print("Models loaded successfully!")


def predict_combined(image_path):

    print("\nReading image...")

    # Check file exists
    if not os.path.exists(image_path):
        print("Image path not found!")
        print(image_path)
        return

    # Read image
    img = cv2.imread(image_path)

    # Check image loaded
    if img is None:
        print("Failed to load image")
        return

    print("Image loaded successfully")
    print("Image Shape:", img.shape)


    img = cv2.resize(img, (128,128))

    img = img / 255.0

    img = np.expand_dims(img, axis=0)


    print("\nPredicting using Flood Model...")

    flood_pred = flood_model.predict(img)

    flood_class = np.argmax(flood_pred)

    flood_classes = ["Flood", "Not Flood"]

    flood_result = flood_classes[flood_class]


    print("\nPredicting using Disaster Model...")

    disaster_pred = disaster_model.predict(img)

    disaster_class = np.argmax(disaster_pred)

    disaster_classes = [
        "Earthquake",
        "Fire",
        "Non-Disaster"
    ]

    disaster_result = disaster_classes[disaster_class]


    if flood_result == "Flood":

        final_result = "Flood"

    elif disaster_result != "Non-Disaster":

        final_result = disaster_result

    else:

        final_result = "Non-Disaster"


    print("\n================================")
    print("       FINAL PREDICTION")
    print("================================")

    print("Flood Model Result      :", flood_result)

    print("Disaster Model Result  :", disaster_result)

    print("\nFinal Prediction :", final_result)


predict_combined(
"/content/drive/MyDrive/major/input/1 (286).jpeg"
)
```
